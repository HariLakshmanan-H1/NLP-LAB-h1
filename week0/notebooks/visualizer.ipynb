```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import numpy as np
```

```python
df_csv = pd.read_csv('../datasets/reviews.csv')
df_json = pd.read_json('../datasets/reviews.json')
df_txt = pd.read_csv('../datasets/reviews.txt', sep='\t')
```

```python
fig = px.pie(df_csv,names="Sentiment_Label",title="Class Distribution")
fig.show()
```

```python
# Code
plt.figure(figsize=(6, 4))
sns.countplot(x='Source', data=df_json, palette='viridis')
plt.title('Count of Reviews by Source')
plt.show()
```

```python

plt.figure(figsize=(6, 4))

sns.barplot(x='Sentiment_Label', y='Word_Count', data=df_txt, palette='muted')
plt.title('Average Review Length by Sentiment')
plt.show()
```

```python

plt.figure(figsize=(6, 4))
sns.scatterplot(x='Word_Count', y='Sentiment_Score', data=df_csv, hue='Sentiment_Label', s=100)
plt.title('Sentiment Score vs. Word Count')
plt.axhline(0.5, color='red', linestyle='--') # Threshold line
plt.show()
```

```python

from sklearn.metrics import confusion_matrix

# Create the matrix
cm = confusion_matrix(df_csv['Actual_Label'], df_csv['Sentiment_Label'])
labels = ['Negative', 'Neutral', 'Positive'] # Alphabetical order usually

plt.figure(figsize=(5, 4))
sns.heatmap(cm, annot=True, xticklabels=labels, yticklabels=labels, cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix Heatmap')
plt.show()
```

```python
sns.pairplot(df_csv, hue='Sentiment_Label', vars=['Sentiment_Score', 'Word_Count'])
plt.show()
```

```python
sns.pairplot(df_csv[['Sentiment_Score', 'Word_Count', 'Source']], hue='Source')
plt.show()
```
